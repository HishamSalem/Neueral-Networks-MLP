## MLP Regressor

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error

#read data into data frame, data
data = pd.read_csv(r'C:\Users\hisha\Desktop\Certfication\winequality.csv')

# Fill missing values with mean column values in the data set
#  In this case, Age had some holes
data.fillna(data.mean(), inplace=True)

#drop the columns that we don't need!
data = data.drop(['pH', 'density', 'sulphates', 'chlorides', 'residual sugar','fixed acidity'], axis=1)

#list all of the predictors that will be used
predictors = ['volatile acidity', 'citric acid', 'free sulfur dioxide','total sulfur dioxide','alcohol' ]

#set up target, predictors, and split the training/testing partitions
X = data[predictors]
y = data['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

#Normalize your data Here!!
scaler = StandardScaler()

# Fit only to the training data
scaler.fit(X_train)
StandardScaler(copy=True, with_mean=True, with_std=True)

# Now apply the transformations to the data:
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python

#####################################################################

print("Model 1 - (15) with relu activation function")
#create the neural network with your choice of hidden layer size
#If your sample size is NOT in the thousands, use solver='lbfgs'
network = MLPRegressor(hidden_layer_sizes=(15),solver='lbfgs')
network.fit(X_train, y_train)

#test the model against our new model and calculate the accuracy
pred = network.predict(X_test)
print('Mean Absolute Error:', mean_absolute_error(y_test, pred))
print()

#####################################################################
```

```python
print("Model 2 - (5,5) with relu activation function")
#create the neural network with your choice of hidden layer size
#If your sample size is NOT in the thousands, use solver='lbfgs'
network = MLPRegressor(hidden_layer_sizes=(5,5),solver='adam',activation='relu')
network.fit(X_train, y_train)

#test the model against our new model and calculate the accuracy
pred = network.predict(X_test)
print('Mean Absolute Error:', mean_absolute_error(y_test, pred))
print()
```
